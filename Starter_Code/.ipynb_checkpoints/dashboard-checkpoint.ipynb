# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/sfo_neighborhoods_census_data.csv")
sfo_data = pd.read_csv(file_path, index_col="year")

file_path = Path("Data/neighborhoods_coordinates.csv")
df_neighborhood_locations = pd.read_csv(file_path)
```

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python
# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""    
    # YOUR CODE HERE!
    sto_mean_units = sfo_data.groupby('year')['housing_units'].mean()
    return sto_mean_units.hvplot(kind='bar', title='Housing Units in San Francisco from 2010 to 2016', ylim=(370000,387500))
```

```python
def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    
    # YOUR CODE HERE!
    return sfo_data.groupby('year')['gross_rent'].mean().hvplot(kind='line', title='Average Gross Rent in San Francisco')
```

```python
def average_sales_price():
    """Average Sales Price Per Year."""    
    # YOUR CODE HERE!
    return sfo_data.groupby('year')['sale_price_sqr_foot'].mean().hvplot(kind='line', title='Average Sale Price per Square Foot in San Francisco')
    sfo_data.set_xlabel("Year")
    sfo_data.set_ylabel("Avg. Sale Price")
```

```python
def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""
    
    # YOUR CODE HERE!
    sto_mean_neighborhoods = sfo_data.groupby(['year','neighborhood'])['sale_price_sqr_foot','housing_units','gross_rent'].mean()
    list_of_hoods = sfo_data['neighborhood'].to_list()
    list_of_hoods = list(dict.fromkeys(list_of_hoods))
    return sto_mean_neighborhoods.hvplot.line(
        x='year',
        y=['sale_price_sqr_foot'],
        groupby=['neighborhood'])
```

```python
def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""
    
    # YOUR CODE HERE!
    sto_most_expensive_neighborhoods = sfo_data.groupby('neighborhood')['sale_price_sqr_foot','housing_units','gross_rent'].mean().sort_values(by=['sale_price_sqr_foot'], ascending=False)
    sfo_expesnive = sto_most_expensive_neighborhoods.iloc[0:10, 0:1]
    return sfo_expesnive.hvplot(kind='bar', title='Average Sale Price Most Expensive')
    sfo_expesnive.set_xlabel("Neighborhood")
    sfo_expesnive.set_ylabel("Avg. Sale Price")
```

```python
def parallel_coordinates():
    """Parallel Coordinates Plot."""
    
    # YOUR CODE HERE!
    sto_most_expensive_neighborhoods = sfo_data.groupby('neighborhood')['sale_price_sqr_foot','housing_units','gross_rent'].mean().sort_values(by=['sale_price_sqr_foot'], ascending=False)
    sfo_top_ten = sto_most_expensive_neighborhoods.iloc[0:10].reset_index()
    return px.parallel_coordinates(sfo_top_ten, color='sale_price_sqr_foot')
```

```python
def parallel_categories():
    """Parallel Categories Plot."""
    # YOUR CODE HERE!
    sto_most_expensive_neighborhoods = sfo_data.groupby('neighborhood')['sale_price_sqr_foot','housing_units','gross_rent'].mean().sort_values(by=['sale_price_sqr_foot'], ascending=False)
    sfo_top_ten = sto_most_expensive_neighborhoods.iloc[0:10].reset_index()
    return px.parallel_categories(
        sfo_top_ten, 
        color='sale_price_sqr_foot', 
        color_continuous_scale=px.colors.sequential.Inferno,)
```

```python
def neighborhood_map():
   # """Neighborhood Map"""
    # YOUR CODE HERE!
    sfo_mean_by_hood = sfo_data.groupby(['neighborhood'])['sale_price_sqr_foot','housing_units','gross_rent'].mean().reset_index()
    joi_df = pd.concat([df_neighborhood_locations, sfo_mean_by_hood], axis="columns", join="inner")
    join_df = joi_df.drop(columns=['neighborhood'])
    return px.scatter_mapbox(join_df, lat="Lat", lon="Lon", color="gross_rent", size='sale_price_sqr_foot', title="Average Sale Price Per Square Foot and Gross Rent in San Francisco")
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# YOUR CODE HERE!
tabs = pn.Tabs(
    ("Yearly Housing Units", housing_units_per_year),
    ("Average Rent", average_gross_rent),
     ("Average Sale Price", average_sales_price),
     ("Price by Hood", average_price_by_neighborhood),
     ("Most Expensive", top_most_expensive_neighborhoods),
     ("Most Expensive Part Duex", parallel_coordinates),
     ("Most Expensive Part Trois", parallel_categories),
     ("San Francisco", neighborhood_map())
)
tabs
```
